<h2>Import Libraries

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
import cv2
import matplotlib.pyplot as plt
```

<h2>Functions

```python
def load_and_prepare_image(img_path, img_size=(150, 150)):
    """
    Load an image and prepare it for prediction by resizing and scaling.

    Args:
    img_path (str): Path to the image file.
    img_size (tuple): Target size for the image (width, height).

    Returns:
    np.array: Preprocessed image ready for model prediction.
    """
    # Load the image
    img = image.load_img(img_path, target_size=img_size)
    
    # Convert the image to a numpy array
    img_array = image.img_to_array(img)
    
    # Scale the image
    img_array = img_array / 255.0
    
    # Expand dimensions to match the shape required by the model
    img_array = np.expand_dims(img_array, axis=0)
    
    return img_array, img

def predict_image_class(model, img_path, class_names):
    """
    Predict the class of an image using a pre-trained model.

    Args:
    model (tf.keras.Model): The pre-trained Keras model.
    img_path (str): Path to the image file.
    class_names (list): List of class names.

    Returns:
    str: Predicted class name.
    """
    # Load and prepare the image
    img_array, original_img = load_and_prepare_image(img_path)
    
    # Make a prediction
    predictions = model.predict(img_array)
    
    # Get the predicted class
    predicted_class = np.argmax(predictions[0])
    
    # Get the class name
    predicted_class_name = class_names[predicted_class]
    
    # Display the image and prediction
    plt.imshow(original_img)
    plt.title(f'Predicted class: {predicted_class_name}')
    plt.axis('off')
    plt.show()
    
    return predicted_class_name
```

<h2>Load Model

```python
model = tf.keras.models.load_model('cnn_flower_classification_model.h5')
```

<h2>Test With Image

```python
img_path = 'sunflower.jpg'
predicted_class_name = predict_image_class(model, img_path, ["Lily", "Lotus", "Orchid", "Sunflower", "Tulip"])
```
